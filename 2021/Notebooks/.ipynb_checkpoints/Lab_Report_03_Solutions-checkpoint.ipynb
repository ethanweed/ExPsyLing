```python
import pandas as pd
import seaborn as sns
import pingouin as png
```

```python
raw = pd.read_csv("https://raw.githubusercontent.com/ethanweed/ExPsyLing/master/datasets/Sentence-verification/2021/sentence-verification.csv")
```

First let's look at the column headers and pick what we need

```python
print(list(raw))
```

Since we can't use `jatosStudyResultId` for our participant numbers like last time (I assume because the server crashed halfway through), let's check to see if `sessionid` gives us the right number of participants

```python
len(set(list(raw['sessionid'])))
```

24 seems about right, so we'll use that. Now we can build a dataframe with the data that we need.

```python


df = pd.DataFrame(
    {'participantID': raw['sessionid'],
     'rt': raw['response_time_stim'],
     'true_false': raw['true_false'],
     'condition': raw['condition'],
     'sentence': raw['stim'],
     'correct_response': raw['correct_response'],
     'response': raw['response_stim']
    })

df
```

We can figure out whether people answerd correctly or not by comparing the actual response with the correct response:

```python

df['correct'] = (df['correct_response'] == df['response']).astype(int)
df
```

```python
# check how many correct answers there were by first dropping the duplicate rows and then looking the "shape"
# of the resulting dataframe
df = df.drop_duplicates()
df[df['correct'] == 1].shape
```

If we want, we can now filter out the wrong answers:

```python
df = df[df['correct'] == True]
```

```python
sns.pointplot(data = df, x = 'condition', y = 'rt', hue = 'true_false')
```

Now we need to remove that pesky "S2." and change the "O"s to "0"s

```python
# change "S2." to "S2"

condition = []

for x in list(df['condition']):
    if x == 'S2.':
        condition.append('S2')
    else:
        condition.append(x)

df['condition'] = condition


# Change "PO" to "P0"
condition = []
for x in list(df['condition']):
    if x == 'PO':
        condition.append('P0')
    else:
        condition.append(x)
        
df['condition'] = condition


# Change "SO" to "S0"
condition = []
for x in list(df['condition']):
    if x == 'SO':
        condition.append('S0')
    else:
        condition.append(x)
        
df['condition'] = condition


# Put the conditions in numeric order
df = df.sort_values('condition')
```

```python
sns.pointplot(data = df, x = 'condition', y = 'rt', hue = 'true_false')
```

```python
df.head()
```

Then we can separate our the P's from the S's:

```python
p_data = df.loc[((df['condition'] == 'P0') | (df['condition'] == 'P1') | (df['condition'] == 'P2'))]

s_data = df.loc[((df['condition'] == 'S0') | (df['condition'] == 'S1') | (df['condition'] == 'S2'))]
```

... and add a new column saying which category the data belong to:

```python
p_data['type'] = ['property']*p_data.shape[0]

level = []

for i in list(p_data['condition']):
    if i == 'P0':
        level.append('0')
    elif i == 'P1':
        level.append('1')
    elif i == 'P2':
        level.append('2')
p_data['level'] = level

s_data['type'] = ['superset']*s_data.shape[0]

level = []
for i in list(s_data['condition']):
    if i == 'S0':
        level.append('0')
    elif i == 'S1':
        level.append('1')
    elif i == 'S2':
        level.append('2')
s_data['level'] = level


# don't worry about the error messages. 
```

```python
p_data
```

```python
s_data
```

Now that we have separted the conditions into two different columns ("type" and "level"), we can recombine our dataframes:

```python
df = pd.concat([p_data, s_data])
df.head()
```

and now we should be set up to run whatever analyses we want. E.g. we can plot our data in a more sensible way:

```python
sns.pointplot(data = df, x = 'level', y = 'rt', hue = 'type')
```

and run e.g. a 2-way repeated measures ANOVA:

```python
df.rm_anova(dv = 'rt', within = ['type', 'level'], subject = 'participantID')
```

and, if we want, post-hoc t-tests:

```python
df.pairwise_ttests(dv = 'rt', within = ['type', 'level'], subject = 'participantID', padjust = 'holm')
```
