```python


days = []
temp = ""
while temp != "done":
    temp = input("Enter a day name. If you are done, write \"done\".")
    if temp != "done":
        days.append(temp)


meals = []
for day in days:
    meals.append(input("What would you like to make on " + day + "? "))
    


```

```python
days
```

```python
meals
```

```python
meals[2]
```

```python
import pandas as pd

df = pd.DataFrame(
    {'days': days,
     'meals': meals
    })

df
```

```python
shopping = []
for meal in df['meals']:
    shopping.append(input("What do you need to buy to make " + meal + " ? "))

df['shopping'] = shopping
```

```python
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_albums")
```

```python
rawdata[1]
```

```python
rawdata[1].shape
```

```python
data = pd.concat([rawdata[1], rawdata[2], rawdata[3]])
data.head()
```

```python
data.shape
```

```python
# make a new column with numbers from 1 to the total number of rows
data['rank'] = [n for n in range(1, data.shape[0]+1)]
data.head()
```

```python
data.tail(10)
```

```python
n = 81
list(data.loc[data['rank'] == n]['Artist'])
```

```python
n = 24
data.loc[data['rank'] == n]
```

```python
data = data[ ['rank'] + [ col for col in data.columns if col != 'rank' ] ]
```

```python
data
```

```python
# Goodby "Ref(s)"
data = data.loc[:,'Artist':'Claimed sales*']
data.head()
```

```python
# get a list of all the column names
cols = list(data.columns)
cols
```

```python
# "pop" out the fourth one ("Total copies")
cols.pop(4)
```

```python
cols
```

```python
# make a new dataframe containing only the columns in the list `cols`
data = data[cols]
data
```

```python
data_sorted = data.sort_values(by=['Artist'])
data_sorted
```

```python
data = data.rename(columns={'Claimed sales*': 'Sales'})
data.head(2)
```

```python
data = data.sort_values(by=['Artist', 'Album'])
data
```

```python
pop = []
for i in list(data['Genre']):
    if 'Pop' in i:
        pop.append(1)
    elif 'pop' in i:
            pop.append(1)
    else:
        pop.append(0)
data['Pop'] = pop
data
```
