```python
# make a new dataframe df, with the essential data from the raw data
df = pd.DataFrame({'participantID' : raw['sessionid'],
                  'condition' : raw['condition'],
                  'sentence' : raw['stim'],
                  'rt' : raw['response_time'],
                   'true_false' : raw['true_false'],
                  'correct' : raw['correct']})

# it turns out that one of the conditions "S2" was mislabeld with a dot: "S2.", so we need to fix that.
condition = []
for x in list(df['condition']):
    if x == 'S2.':
        condition.append('S2')
    else:
        condition.append(x)

df['condition'] = condition


condition = []
for x in list(df['condition']):
    if x == 'PO':
        condition.append('P0')
    else:
        condition.append(x)
        
df['condition'] = condition

condition = []
for x in list(df['condition']):
    if x == 'SO':
        condition.append('S0')
    else:
        condition.append(x)
        
df['condition'] = condition
```

```python
p_data['type'] = ['property']*p_data.shape[0]
level = []
for i in list(p_data['condition']):
    if i == 'P0':
        level.append('0')
    elif i == 'P1':
        level.append('1')
    elif i == 'P2':
        level.append('2')
p_data['level'] = level
p_data
```

```python
s_data['type'] = ['superset']*s_data.shape[0]

level = []
for i in list(s_data['condition']):
    if i == 'S0':
        level.append('0')
    elif i == 'S1':
        level.append('1')
    elif i == 'S2':
        level.append('2')
s_data['level'] = level
s_data

s_data
```
