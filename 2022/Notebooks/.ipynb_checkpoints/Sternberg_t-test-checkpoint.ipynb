```python
import pandas as pd
import numpy as np
import pingouin as pg
import seaborn as sns
from matplotlib import pyplot as plt

sns.set_context("notebook", font_scale=1.5)

file = 'https://raw.githubusercontent.com/ethanweed/ExPsyLing/master/datasets/Sternberg/2021/Sternberg_cleaned.csv'

data = pd.read_csv(file)
```

```python
data.head()
```

```python
df_p = data[data['present'] == "y"]
df_n = data[data['present'] == "n"]

df = pd.DataFrame({'present': df_p.groupby('id').mean()['rt'],
                  'not_present': df_n.groupby('id').mean()['rt']})

df.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

sns.kdeplot(data = df['present'], ax=ax, fill=True, label='present')
sns.kdeplot(data = df['not_present'], ax=ax, fill=True, label='not present')

plt.axvline(np.mean(df['present']), color = 'blue')
plt.axvline(np.mean(df['not_present']), color = 'red')


plt.axvline(np.mean(df['present'])-(np.std(df['present'])*2), color = 'blue', linestyle = '--')
plt.axvline(np.mean(df['present'])+(np.std(df['present'])*2), color = 'blue', linestyle = '--')

plt.axvline(np.mean(df['not_present'])-(np.std(df['not_present'])*2), color = 'red', linestyle = '--')
plt.axvline(np.mean(df['not_present'])+(np.std(df['not_present'])*2), color = 'red', linestyle = '--')


sns.despine()
```

```python
pg.ttest(df['present'], df['not_present'], paired = True, alternative = 'two-sided')
```

```python
sns.pointplot(data = df)
```

```python
sample_size = 30

present = []
not_present = []

number_of_experiments = 10
for i in range(number_of_experiments):
    present.append(np.random.normal(np.mean(df['present']), np.std(df['present']), sample_size))
    not_present.append(np.random.normal(np.mean(df['not_present']), np.std(df['not_present']), sample_size))
```

```python
for i in range(len(present)):
    sns.kdeplot(data = present[i], color = "red")
    sns.kdeplot(data = not_present[i], color = "blue")
```

```python
t_values = []
p_values = []
for i in range(len(present)):
    sim = pg.ttest(present[i], not_present[i], paired = True, alternative = 'two-sided')
    t = sim['T'].tolist()[0]
    t_values.append(t)
    p = sim['p-val'].tolist()[0]
    p_values.append(p)
```

```python
sns.kdeplot(data = t_values)
plt.axvline(np.mean(t_values), color = 'black', linestyle = '--')
```

```python
sns.histplot(data = p_values)
plt.axvline(np.mean(p_values), color = 'black', linestyle = '--')
```
