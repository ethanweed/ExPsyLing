```python
df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/ExPsyLing/master/datasets/StudentSleep.csv")
df
```

```python
# find the number of rows and columns of the datafrme
df.shape
```

```python
# select the first four rows of the dataframe
df1 = df.iloc[0:4]
df1
```

```python
# select the last 3 rows of the dataframe
df2 = df.iloc[4:7]
df2
```

```python
 # another way to select the last 3 rows of the dataframe
df2 = df.iloc[4:]
df2
```

```python
 # another way to select the last 3 rows of the dataframe
df2 = df.iloc[4:df.shape[0]]
df2
```

```python
# stack dataframe 2 on top of dataframe 1
# notice that pandas keeps track of the index number (row number)
df3 = pd.concat([df2, df1])
df3
```

```python
# stack dataframe 1 on top of dataframe 2
df3 = pd.concat([df1, df2])
df3
```

```python
# create a new column with the row means for all the other columns
df3['average'] = df3.mean(axis = 1)
df3
```

```python
# remove a column
df3.pop('average')
df3
```

```python
# add a new row with means of all columns (not so useful!)
df3.loc['mean'] = df3.mean()
df3
```

```python
# make a list variable with means of all columns (more useful!)
column_means = list(df3.mean())
column_means
```

```python
# here is our original data again
df
```

```python
# here we flip the dataframe on its side (transpose it)
df_transposed = df.transpose()
df_transposed
```

```python
# get a list of all the column names
colnames = list(df_transposed)
colnames
```

```python
# use list comprehensions to rename the columns in a more readable way
newcols = ['Day_' + str(x+1) for x in colnames]
newcols
```

```python
# replace the column names with the new list we made
df_transposed.columns = newcols
df_transposed
```

```python
# make a new column using the "indices" from the old column, that is:
# make a column caled "students"
# we will need this so that we can "melt" the data from wide to long format
df_transposed.index.name = 'student'
df_transposed.reset_index(inplace = True)
df_transposed
```

```python
# melt the column from wide to long format
df_long = pd.melt(df_transposed, id_vars = 'student')
df_long
```
