```python
df = df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/ExPsyLing/master/datasets/StudentSleep.csv")
df
```

```python
# index and columns
```

```python
df.shape
```

```python
df1 = df.iloc[0:4]
df1
```

```python
# NOTICE THE 4 IS DIFFERENT WHEN IT IS THE START AND WHEN IT IS THE END!
df2 = df.iloc[4:7]
df2
```

```python
# ANOTHER WAY TO ACHIEVE THE SAME THING
df2 = df.iloc[4:df.shape[0]]
df2
```

```python
# CONCATENATE

# NOTICE THE INDICES

df3 = pd.concat([df2, df1])
df3
```

```python
df4 = pd.concat([df1,df2])
df4
```

```python
df4['average'] = df4.mean(axis = 1)
df4
```

```python
df4.pop('average')
df4
```

```python
df4.loc['mean'] = df4.mean()
df4
```

```python
column_means = list(df4.mean())
column_means
```

```python
column_means = [round(x,3) for x in column_means]
column_means
```

```python
first_three = list(df)[0:3]
first_three
```

```python
last_three = list(df)[-3:]
last_three
```

```python
df1 = df[first_three]
df1
```

```python
df2 = df[last_three]
df2
```

```python
df3 = df1.join(df2)
df3
```

```python
df_transposed = df3.transpose()
df_transposed
```

```python
colnames = list(df_transposed)
colnames
```

```python
colnames = ['Day_' + str(x) for x in colnames]
colnames
```

```python
df_transposed.columns = colnames
df_transposed
```

```python
df_transposed.index.name = 'student'
df_transposed.reset_index(inplace = True)
df_transposed
```

```python
df_long = pd.melt(df_transposed, id_vars = 'student')
df_long
```

```python
df_long.columns = ['student', 'day', 'sleep']
df_long
```

```python
df_wide = pd.pivot(df_long, index=['student'], columns='day')
df_wide
```

```python
list(df_wide)
```

```python
df_wide.columns = [x[1] for x in list(df_wide)]
df_wide
```
