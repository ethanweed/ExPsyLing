```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

file = "https://raw.githubusercontent.com/ethanweed/ExPsyLing/master/Experiments/LexicalDecision/LexicalDecision_1/data/Lab_report_01.csv"
```

Here you can see the whole link:

https://raw.githubusercontent.com/ethanweed/ExPsyLing/master/Experiments/LexicalDecision/LexicalDecision_1/data/Lab_report_01.csv

```python
rawdata = pd.read_csv(file)
```

```python
rawdata.head()
```

```python
len(rawdata)
```

```python
rawdata.shape
```

```python
df = rawdata[(rawdata['block'] == 'trial')]
```

```python
len(df)
```

```python
df.head()
```

```python
set(df['block'])
```

```python
df_means = df.groupby(['jatosStudyResultId', 'category', 'number_letters']).mean()
```

```python
df_means.shape
```

```python
df_means.head(10)
```

```python
df_means.reset_index(inplace = True)
df_means
```

```python
df_means.reset_index(inplace = True)
df_means['id'] = df_means['jatosStudyResultId']

data = df_means[['id', 'category', 'accuracy', 'number_letters', 'avg_rt']]
data.head()
```

```python
data.shape
```

```python
len(set(data['id']))
```

## Quick look at the data

```python
ax = sns.scatterplot(data = data, x = 'number_letters', y = 'avg_rt')
```

```python
sns.kdeplot(x = 'avg_rt', hue = 'category', data = data)
```

## Data as distributions

- what is a distribution?
- what are density (kde) plots)
- observed data as samples from a distribution
- data are generated by a process (relate to causal models, "error" in e.g. linear models)
- individual distributions, group sample distributions, population distributions (usually unknowable, but we can estimate)

```python
ax = sns.kdeplot(x = 'avg_rt', hue = 'id', data = data)
ax = sns.kdeplot(x = 'avg_rt', data = data)
plt.legend([],[], frameon=False)
```

## Painting with numbers and matplotlib

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

ax1 = sns.kdeplot(x = 'avg_rt', hue = 'id', 
                  data = data[(data['category'] == 'NW')], 
                  ax=axes[0])


ax2 = sns.kdeplot(x = 'avg_rt', hue = 'id', 
                  data = data[(data['category'] == 'PH')], 
                  ax=axes[1])


ax3 = sns.kdeplot(x = 'avg_rt', hue = 'id', 
                  data = data[(data['category'] == 'Word')], 
                  ax=axes[2])

ax2.set(ylabel = '')
ax3.set(ylabel = '')



for ax in axes:
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.get_legend().remove()
    ax.set_xlim(500, 1400)
    ax.set_ylim(0, 0.007)
    
titles = ['Non-words', 'Pseudohomophones', 'Words']

for n, ax in enumerate(axes):
    ax.set_title(titles[n])

```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

ax1 = sns.kdeplot(x = 'accuracy', hue = 'id', 
                  data = data[(data['category'] == 'NW')], 
                  ax=axes[0])

ax1 = sns.kdeplot(x = 'accuracy', data = data[(data['category'] == 'NW')], 
                  ax=axes[0], linewidth = 3)




ax2 = sns.kdeplot(x = 'accuracy', hue = 'id', 
                  data = data[(data['category'] == 'PH')], 
                  ax=axes[1])

ax2 = sns.kdeplot(x = 'accuracy', data = data[(data['category'] == 'PH')], 
                  ax=axes[1], linewidth = 3)


ax3 = sns.kdeplot(x = 'accuracy', hue = 'id', 
                  data = data[(data['category'] == 'Word')], 
                  ax=axes[2])

ax3 = sns.kdeplot(x = 'accuracy', data = data[(data['category'] == 'Word')], 
                  ax=axes[2], linewidth = 3)

                  
ax2.set(ylabel = '')
ax3.set(ylabel = '')

titles = ['Non-words', 'Pseudohomonyms', 'Words']

for ax in axes:
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.get_legend().remove()
    #ax.set_xlim(500, 1400)
    #ax.set_ylim(0, 0.007)


for n, ax in enumerate(axes):
    ax.set_title(titles[n])
    


```
