```python
# use list comprehensions to compare two lists, and select items from list 1 that match items in list 2
b = a[0:20]
favs = ['Barkley',  'Betty Lou', 'Bert', 'Kermit']
c = [x for x in b if x in favs]
c


# here, Kermit was in my list of favorite monsters, but he was not in list b
```

```python
# use list comprehensions to compare two lists, and select items from list 1 that are not in list 2
b = a[0:20]
favs = ['Barkley',  'Betty Lou', 'Bert']
c = [x for x in b if x not in favs]
c

# this gives us all the monters in the list that are NOT my favorites

# Not really. I love all monsters.
```

## dataframe manipulation

1. concatenating dataframes
2. joining dataframes
3. adding new columns
4. popping dataframe columns
5. wide-to-long conversion
6. long-to-wide conversion
7. sorting dataframes

```python
df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/ExPsyLing/master/datasets/StudentSleep.csv")
df
```

```python
df.shape
```

```python
df1 = df.iloc[0:4]
df1
```

```python
df2 = df.iloc[4:df.shape[0]]
df2
```

```python
df3 = pd.concat([df2, df1])
df3
```

```python
df3 = pd.concat([df1, df2])
df3
```

```python
df3['average'] = df3.mean(axis = 1)
df3
```

```python
df3.pop('average')
```

```python
df3
```

```python
column_means = list(df3.mean())
column_means
```

```python
colnames = list(df3)
colnames
```

```python
list(zip(colnames, column_means))
```

```python
df3
```

```python
first_three = list(df)[0:3]
first_three
```

```python
df1 = df3[first_three]
df1
```

```python
last_three = list(df)[-3:]
last_three
```

```python
df2 = df[last_three]
df2
```

```python
df3 = df1.join(df2)
df3
```

```python
df_transposed = df3.transpose()
df_transposed
```

```python
colnames = list(df_transposed)
colnames
```

```python
newcols = ['Day ' + str(x+1) for x in colnames]
newcols
```

```python
df_transposed.columns = newcols
df_transposed
```

```python
df_transposed.index.name = 'student'
df_transposed.reset_index(inplace = True)
df_transposed
```

```python
df_long = pd.melt(df_transposed, id_vars = 'student')
df_long
```
